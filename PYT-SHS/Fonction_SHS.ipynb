```python
def compterMot (text, word):
    #On met tout en minuscules 
    text = text.lower()
    word = word.lower()
    #On découpe le texte en mots
    mots = text.split()
    #On compte les occurrences
    return mots.count(word)
```

```python
#Définir le texte
text = "Le Québec est une terre de contrastes et de richesses. Depuis des siècles, le Québec porte fièrement sa langue française comme un flambeau au cœur de l'Amérique du Nord. Les habitants du Québec ont su bâtir une culture unique, mêlant traditions ancestrales et modernité."

#Définir le mot
word = "Québec"

#Utiliser la fonction avec les arguments
compterMot (text, word)
```
